```python
#Course: Data Science Masters
#Title: Pandas - Assignment 13.1            
#Tool: Jupyter notebook 
#Author: Krishna A
#Date: 07-05-2018 (DD/MM/YYYY)

# SQL Assignments

#Problem Statement :Read the following data set:
#                          https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data
#                   Rename the columns as per the description from this file:
#                          https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.names

import pandas as pd
from pandasql import sqldf

pysqldf = lambda statement: sqldf(statement, globals())

sFile = r'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
sColNames = ['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship',
                   'race','sex','capital_gain','capital_loss','hours_per_week','native_country','class']

dfData = pd.read_csv(sFile, header=None, names = sColNames)

### trim columns ### 
dfCols = dfData.select_dtypes(['object'])
dfData[dfCols.columns] = dfCols.apply(lambda x: x.str.strip())

dfData.head(5)
```

```python
#Problem Statement 1:  1.1 - Create a sql db from adult dataset and name it sqladb
#                      1.2 - Select 10 records from the adult sqladb

sqladb = pysqldf("SELECT * FROM dfData;")
pysqldf('SELECT * FROM sqladb LIMIT 10;')  #Limit keyword restricts data retrieval to set number of records
```

```python
#Problem Statement 2: Show me the average hours per week of all men who are working in private sector

pysqldf("SELECT AVG(hours_per_week) FROM sqladb WHERE sex = 'Male' AND workclass= 'Private';")
```

```python
#Problem Statement 3: Show me the frequency table for education, occupation and relationship, separately

# For education
pysqldf("SELECT education, COUNT(education) as 'education_frequency' FROM sqladb GROUP BY education;")
```

```python
# For occupation
pysqldf("SELECT occupation, COUNT(occupation) as 'occupation_frequency' FROM sqladb GROUP BY occupation;")
```

```python
# For relationship
pysqldf("SELECT relationship, COUNT(relationship) as 'relationship_frequency' FROM sqladb GROUP BY relationship;")
```

```python
#Problem Statement 4:Are there any people who are married, working in private sector and having a masters degree?

pysqldf("SELECT COUNT(*) FROM sqladb WHERE marital_status LIKE 'married%' AND workclass='Private' AND education='Masters';")
```

```python
#Problem Statement 5: What is the average, minimum and maximum age group for people working in
#                     different sectors
pysqldf("""
        SELECT workclass, AVG(age) as 'AverageAgeGroup', MIN(age) as 'MinimumAgeGroup', MAX(age) as 'MaximumAgeGroup'
        FROM sqladb GROUP BY workclass
        """)
```

```python
#Problem Statement 6: Calculate age distribution by country
pysqldf("""
    SELECT 
        native_country,
        COUNT(distinct AGE) as DistinctAgeCount,
        MIN(age) as MinAge,
        MAX(age) as MaxAge,
        AVG(age) as AverageAge
    FROM sqladb
    GROUP BY native_country
    ORDER BY native_country;
""")
```

```python
#Problem Statement 7: Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and
#                    'capital-loss'

pysqldf("""
    SELECT 
        capital_gain, capital_loss,
        (capital_gain-capital_loss) AS 'Net-Capital-Gain'
    FROM sqladb;
""")
```
